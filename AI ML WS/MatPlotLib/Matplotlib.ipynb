## Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graphs

```python
plt.plot([1,2,3],[2,4,6])
```

```python
x = [1,2,3,4,5,6]
y = [2,4,6,8,10,12]

plt.plot(x,y)

plt.show() #this is what we use to display the graph
```

```python
x = [1,2,3,4,5,6]
y = [2,4,6,8,10,12]

plt.plot(x,y)
plt.title("Our First Graph") #adding title
plt.xlabel("X axis") #adding x label
plt.ylabel("Y axis")#adding y label

plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y)
plt.title("Our First Graph", fontdict={'fontname':'Comic Sans MS','fontsize':20}) #adding title
#and changing the font style
plt.xlabel("X axis",fontdict={'fontname':'Comic Sans MS'}) #adding x label with style
plt.ylabel("Y axis")#adding y label

plt.show()
```

```python
# above graph has decimals if we want only integers we can do the following op
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y)
plt.title("Our First Graph", fontdict={'fontname':'Comic Sans MS','fontsize':20}) #adding title
#and changing the font style
plt.xlabel("X axis",fontdict={'fontname':'Comic Sans MS'}) #adding x label with style
plt.ylabel("Y axis")#adding y label

#adding specific values to the graphs 
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,7,8]) # no need to be ordered it will place the elem properly

plt.show()
```

```python
#adding legends
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y,label = '2x') #add the label here
plt.title("Our First Graph", fontdict={'fontname':'Comic Sans MS','fontsize':20}) #adding title
#and changing the font style
plt.xlabel("X axis",fontdict={'fontname':'Comic Sans MS'}) #adding x label with style
plt.ylabel("Y axis")#adding y label

#adding specific values to the graphs 
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,7,8]) # no need to be ordered it will place the elem properly

plt.legend()  #use legend method to display it

plt.show()
```

### Changing parameters of plots

```python
plt.plot(x,y,color='red') #hex values can also be passed
plt.show()
```

```python
#changing line size
plt.plot(x,y,color='red',linewidth=3)
```
