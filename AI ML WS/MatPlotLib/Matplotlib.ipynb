## Using shorthand notations

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#the general way to do the changes is below
#plt.plot(x,y,label='2x',color='red',linewidth=2,marker='.',linestyle='--',markersize=10,markeredgecolor='blue')

#shorhand notations is done using the syntax below
#fmt = '[color][marker][line]'
plt.plot(x,y,'r^--',label = '2x') #its after x,y passed as a string

#r is for red , ^ is the marker and -- idicates dashed line
#refer docs for more supported notations in shorthand form

plt.title("Our First Graph", fontdict={'fontname':'Comic Sans MS','fontsize':20})

plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.show()
```

## graphs using numpy

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y,'b^--',label = '2x') #blue line as b is used

#Line number 2 or second graph
x2 = np.arange(0,4.5,0.5)
plt.plot(x2,x2**2,'r',label='x^2')

plt.title("Our First Graph", fontdict={'fontname':'Comic Sans MS','fontsize':20})

plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.show()
```

## plot different styles in same plot

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y,'b^--',label = '2x')

#Line number 2 or second graph
x2 = np.arange(0,4.5,0.5)
#printing parts of the graph||
plt.plot(x2[:6],x2[:6]**2,'r',label='x^2')# values till 6th index 
plt.plot(x2[5:],x2[5:]**2,'r--') # values after that

plt.title("Our First Graph", fontdict={'fontname':'Comic Sans MS','fontsize':20})

plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.show()
```

### resize the graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#resize the graph using the following method - dpi and figsize properties
plt.figure(figsize=(5,3),dpi=100) #this gives a resolution of 500x300 (5*100)

plt.plot(x,y,'b^--',label = '2x')

#Line number 2 or second graph
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,'r',label='x^2')
plt.plot(x2[5:],x2[5:]**2,'r--')

plt.title("Our First Graph", fontdict={'fontname':'Comic Sans MS','fontsize':20})

plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.show()
```

## Saving the graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.figure(figsize=(5,3),dpi=100) 

plt.plot(x,y,'b^--',label = '2x')

#Line number 2 or second graph
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,'r',label='x^2')
plt.plot(x2[5:],x2[5:]**2,'r--')

plt.title("Our First Graph", fontdict={'fontname':'Comic Sans MS','fontsize':20})

plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])

plt.legend()

#use the below method to save the graph as an image
plt.savefig('mygraph.png',dpi=300) #we can change resolution while we save

plt.show()
```

# BARPLOTS

```python
labels = ['A','B','C']
values = [1,4,2]

plt.bar(labels,values)

plt.show()
```

```python
#all properties are same as previous plt like labels,ticks,colors etc
#resizing is also same and shown below
labels = ['A','B','C']
values = [1,4,2]

plt.bar(labels,values)

plt.figure(figsize=(6,4))

plt.show()
```

```python
#changing the type of design in bars
labels = ['A','B','C']
values = [1,4,2]

bars = plt.bar(labels,values) #assign the plot to bars

bars[0].set_hatch('/')
bars[1].set_hatch('O')
bars[2].set_hatch('*') #pass different designs 

plt.figure(figsize=(6,4))

plt.show()
```

```python
# for cleaning the above code up we can use the below method

labels = ['A','B','C']
values = [1,4,2]

bars = plt.bar(labels,values)

patterns=['/','O','*']

for bar in bars:
    bar.set_hatch(patterns.pop(0)) #using for loop 

plt.figure(figsize=(6,4))

plt.show()
```

### Line Graph

```python
#import matplotlib.pyplot as plt
#import numpy as np
#import pandas as pd
```

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5))

plt.title('Gas Prices Over Time (in USD)')

plt.plot(gas.Year,gas.USA,'b.-',label='USA')
plt.plot(gas.Year,gas.Canada,'r.-',label='Canada')
plt.plot(gas.Year,gas['South Korea'],'g.-',label='South Korea')
plt.plot(gas.Year,gas.Australia,'y.-')

plt.xticks(gas.Year[::3])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()

plt.show()
```

```python
#plotting using for loop to plot all countries
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5))

plt.title('Gas Prices Over Time (in USD)',fontdict={'fontweight':'bold','fontsize':'18'})

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year,gas[country],marker='.',label=country)
        
plt.xticks(gas.Year[::3].tolist()+[2011])  #to add additional years to this for future

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend(bbox_to_anchor=(1,0.5),loc='center left') 
#pass param to move legend outside the graph 
#in this case it is outside to center and left of graph by 1,0.5

plt.savefig('Gas_price_figure.png',dpi=300)

plt.show()
```
