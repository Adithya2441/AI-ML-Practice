## Load Necessary Libraries

### Basic Graphs

```python
plt.plot([1,2,3],[2,4,6])
```

```python
x = [1,2,3,4,5,6]
y = [2,4,6,8,10,12]

plt.plot(x,y)

plt.show() #this is what we use to display the graph
```

```python
x = [1,2,3,4,5,6]
y = [2,4,6,8,10,12]

plt.plot(x,y)
plt.title("Our First Graph") #adding title
plt.xlabel("X axis") #adding x label
plt.ylabel("Y axis")#adding y label

plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y)
plt.title("Our First Graph", fontdict={'fontname':'Comic Sans MS','fontsize':20}) #adding title
#and changing the font style
plt.xlabel("X axis",fontdict={'fontname':'Comic Sans MS'}) #adding x label with style
plt.ylabel("Y axis")#adding y label

plt.show()
```

```python
# above graph has decimals if we want only integers we can do the following op
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y)
plt.title("Our First Graph", fontdict={'fontname':'Comic Sans MS','fontsize':20}) #adding title
#and changing the font style
plt.xlabel("X axis",fontdict={'fontname':'Comic Sans MS'}) #adding x label with style
plt.ylabel("Y axis")#adding y label

#adding specific values to the graphs 
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,7,8]) # no need to be ordered it will place the elem properly

plt.show()
```

```python
#adding legends
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y,label = '2x') #add the label here
plt.title("Our First Graph", fontdict={'fontname':'Comic Sans MS','fontsize':20}) #adding title
#and changing the font style
plt.xlabel("X axis",fontdict={'fontname':'Comic Sans MS'}) #adding x label with style
plt.ylabel("Y axis")#adding y label

#adding specific values to the graphs 
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,7,8]) # no need to be ordered it will place the elem properly

plt.legend()  #use legend method to display it

plt.show()
```

### Changing parameters of plots

```python
plt.plot(x,y,color='red') #hex values can also be passed
plt.show()
```

```python
#changing line size
plt.plot(x,y,color='red',linewidth=3)
```

```python
#some other ops like markers and linestyles and their properties
plt.plot(x,y,color='red',linewidth=2,marker='.',linestyle='--',markersize=10,markeredgecolor='blue')
```

## Using shorthand notations

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#the general way to do the changes is below
#plt.plot(x,y,label='2x',color='red',linewidth=2,marker='.',linestyle='--',markersize=10,markeredgecolor='blue')

#shorhand notations is done using the syntax below
#fmt = '[color][marker][line]'
plt.plot(x,y,'r^--',label = '2x') #its after x,y passed as a string

#r is for red , ^ is the marker and -- idicates dashed line
#refer docs for more supported notations in shorthand form

plt.title("Our First Graph", fontdict={'fontname':'Comic Sans MS','fontsize':20})

plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.show()
```

## graphs using numpy

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y,'b^--',label = '2x') #blue line as b is used

#Line number 2 or second graph
x2 = np.arange(0,4.5,0.5)
plt.plot(x2,x2**2,'r',label='x^2')

plt.title("Our First Graph", fontdict={'fontname':'Comic Sans MS','fontsize':20})

plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.show()
```

## plot different styles in same plot

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y,'b^--',label = '2x')

#Line number 2 or second graph
x2 = np.arange(0,4.5,0.5)
#printing parts of the graph||
plt.plot(x2[:6],x2[:6]**2,'r',label='x^2')# values till 6th index 
plt.plot(x2[5:],x2[5:]**2,'r--') # values after that

plt.title("Our First Graph", fontdict={'fontname':'Comic Sans MS','fontsize':20})

plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.show()
```

### resize the graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#resize the graph using the following method - dpi and figsize properties
plt.figure(figsize=(5,3),dpi=100) #this gives a resolution of 500x300 (5*100)

plt.plot(x,y,'b^--',label = '2x')

#Line number 2 or second graph
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,'r',label='x^2')
plt.plot(x2[5:],x2[5:]**2,'r--')

plt.title("Our First Graph", fontdict={'fontname':'Comic Sans MS','fontsize':20})

plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.show()
```

## Saving the graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.figure(figsize=(5,3),dpi=100) 

plt.plot(x,y,'b^--',label = '2x')

#Line number 2 or second graph
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,'r',label='x^2')
plt.plot(x2[5:],x2[5:]**2,'r--')

plt.title("Our First Graph", fontdict={'fontname':'Comic Sans MS','fontsize':20})

plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])

plt.legend()

#use the below method to save the graph as an image
plt.savefig('mygraph.png',dpi=300) #we can change resolution while we save

plt.show()
```
