# Filtering Data

```python
df.loc[(df['Type 1']=='Grass') & (df['Type 2']=='Poison')] #multiple conditions
```

```python
df.loc[(df['Type 1']=='Grass') | (df['Type 2']=='Poison')] #or operator
```

```python
df.loc[(df['Type 1']=='Grass') & (df['Type 2']=='Poison') & (df['HP']>70)] #conditional operators
```

```python
new_df = df.loc[(df['Type 1']=='Grass') & (df['Type 2']=='Poison') & (df['HP']>70)]
```

```python
new_df
```

```python
#save this using above methods
new_df.to_csv('filtered.csv',index=False)
```

```python
# to reset the index of the data and to remove all the old indexes
#and to replace it in place rather than storing it in a new variable

new_df.reset_index(drop=True,inplace=True)

new_df
```

### REGEX Filtering

```python
df.loc[df['Name'].str.contains('Mega')] #Name containing Mega
```

```python
#the not part of the above command can be done using ~
df.loc[~df['Name'].str.contains('Mega')]
```

```python
#using re

import re

df.loc[df['Type 1'].str.contains('Fire|Grass',regex=True)]
```

```python
#for ignoring case sensitivity


df.loc[df['Type 1'].str.contains('fire|grass',flags=re.I,regex=True)]
```

```python
#starting with specific letters

df.loc[df['Name'].str.contains('^pi[a-z]*',flags=re.I,regex=True)]
```

## Aggregate Statistics (Groupby)

```python
df = pd.read_csv('modified.csv')
```

```python
df
```

```python
df.groupby(['Type 1']).mean().sort_values('HP',ascending=False)
#Grouping the data into type 1 and finding means
```

```python
df.groupby(['Type 1']).sum() #sum of all the data type 1 wise - not very useful for this data
```

```python
df.groupby(['Type 1']).count() #displays count in a wierd way
```

```python
df['count'] = 1

df.groupby(['Type 1']).count()['count'] #only the counts in visible now
```

```python
#groupby multiple cols
df.groupby(['Type 1','Type 2']).count()['count']
```

```python
#These aggregates are really useful in handling the very large datasets
```

## Working with large amounts of data

```python
#instead of loading the entire data we can load chunks of data
for df in pd.read_csv('modified.csv',chunksize=5):
    print("Chunk DF")
    print(df)
```

```python
#make new df using large df with only the required data

new_df = pd.DataFrame(columns=df.columns)

for df in pd.read_csv('modified.csv',chunksize=5):
    results = df.groupby('Type 1').count()
    
    new_df = pd.concat([new_df,results])
```

```python
new_df #counts of Type 1
```
