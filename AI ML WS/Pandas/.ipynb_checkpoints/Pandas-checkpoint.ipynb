### Sorting

```python
df.sort_values('Name')
```

```python
#ascending and descending
```

```python
df.sort_values('Name',ascending=False) #descending order
```

```python
df.sort_values(['Type 1','HP'],ascending=[1,0]) #ascending and descending seperately column wise
```

### Making changes to data

```python
df['Total'] = df['HP']+df['Attack']+df['Defense']+df['Sp. Atk']+df['Sp. Def']+df['Speed']
```

```python
df.head(5) #total of all stats is added to the data as a seperate col
```

```python
#Dropping columns
df = df.drop(columns = ['Total'])
```

```python
df.head(5)
```

```python
#adding col in a diff way
df['Total'] = df.iloc[:, 4:10].sum(axis=1)
```

```python
df.head(5)
```

```python
#relocating col
```

```python
cols = list(df.columns) #array of column names
```

```python
cols
```

```python
df = df[cols[0:4]+[cols[-1]]+cols[4:12]] #single cols shud be b=made to list while concatenating
df.head(5)
```

## Saving this new CSV

```python
df.to_csv('modified.csv',index=False) #index is made false to remove the first col 
```

## Saving the data to excel

```python
df.to_excel('modified.xlsx',index=False)
```

## Saving to txt

```python
df.to_csv('modified.txt',index=False,sep='\t')
```
